```python
#importando os módulos necessários...

import math
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#definindo a equação diferencial

def EqDif(Sol, t): #recebe parâmetros iniciais e lista de tempo
    Theta=Sol[0]
    W=Sol[1]
    dThetadt=W
    dWdt=-g*math.sin(Theta)/r
    return [dThetadt, dWdt] #retorna velocidade do ângulo theta e aceleração do ângulo theta
```

```python
lista_t=np.arange(0,5,0.01) #definindo lista de tempo para simulação (até 5 segundos com intervalos de 10 milissegundos)
r=1.65 #raio do fio: 1,65 metro
condinit= [math.radians(65), 0] #definindo a condição inicial: ângulo inicial de 65 graus, velocidade inicial 0 m/s
g=9.8 #aceleração da gravidade
m=0.37 #massa da pruma: 370 gramas ou 0,37 kg

sol=odeint(EqDif, condinit, lista_t) #chamando o odeint para resolver a equação diferencial
```

```python
X=[] #lista da posição X
Y=[] #lista da posição Y
T=[] #lista da tração
periodo=[] #lista do período do pêndulo

for i in range(len(sol[:,0])): #calculando tração, posição X e Y
    Theta=(sol[:,0])[i]
    W=(sol[:,1])[i]
    X.append(-r*math.cos(Theta))
    Y.append(r*math.sin(Theta))
    T.append((m*(W**2)*r)-(m*9.8*math.cos(Theta)))
    
for h in range(2, len(X)): #calculando período
    if X[1]<=X[h]:
        periodo.append(lista_t[h])
        break
        
print(periodo) #imprimindo período em segundos do pêndulo
```

```python
#plotando para 65 graus

plt.plot(Y, X, 'black')
plt.grid(True)
plt.axis('Equal')
plt.xlabel('Posição em X')
plt.ylabel('Posição em Y')
plt.title('Movimento do Pêndulo')
plt.show()

plt.plot(lista_t, X)
plt.grid(True)
plt.ylabel('Posição em X')
plt.xlabel('Tempo [s]')
plt.title('Posição X em função do tempo')
plt.show()

plt.plot(lista_t, Y)
plt.grid(True)
plt.ylabel('Posição em Y')
plt.xlabel('Tempo [s]')
plt.title('Posição Y em função do tempo')
plt.show()
```
